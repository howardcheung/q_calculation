# Introduction

This IPython Notebook demonstrate how the functions in this folder can be used to analyze the water-side heat exchanger performance by extracting multiple steady state operation data points from time-series data and getting various samples from the data point for its performance at different steady state conditions.

# Initialization

First, we want to import the modules from the customized Python files as

```python
import data_read as dread
import data_summary as ds
import misc_func as misc
```

data_read.py contains functions to read and extract steady state data from uncersteady time-series data to create multiple DataFrames with each represent one steady state data point.

data_summary.py contains functions to solve the heat transfer rate and summarize the data in each steady state data point. 

misc_func.py contains functions and class definitions that help to run the other two python files.

Now let's define the parameters in this analysis. First, we set the file name as a variable as:

```python
filepath = './test_data.csv'
```

In this demonstration, the data are tested under undercharged conditions with undercharged level at 10%. Since the information is not included in the data file, we define it sepearately as

```python
fault_type = 'UC'
fault_level = '10'
```

We will look at how the information can be injected into our summary files generated at the end of this demonstration later.

Now let's define other variables that we will use in this analysis. To begin with, we want our uncertainties from statistical analysis to be in 95% confidence interval. We define

```python
alpha = 0.95
```

We also want the steady state period to be 15 minutes long. 

```python
ss_time = 900
```

There are multiple heat exchangers in the dataset, but we only want the data from one of them. Let's give it an identifier to avoid confusion.

```python
hx_name = 'evap'
```

The data are stored under different headers meaning data obtained from different locations in the experimental setup. Let's define them separately:

```python
tei_name = "TEI" # column name of inlet temperature in degree F
teo_name = "TEO" # column name of outlet temperature in degree F
# column name of temperature of volumetric airflow in degree F
tvdot_name = "TBO"
vdot_name = "FWE"  # column name of flow rate in GPM
kW_name = "kW"  # column name of power consumption in kW
tset_name = "TWE_set"  # column name of setpoint temperature in degree F
# column names of other important temperature measurement in degree F
other_t_names = ["TSI", "TCI", "TCO"] 
# column names of other important flow rates in GPM
other_vdot_names = ["FWC"]
```

The column name indicating the time of data acqusition also needs to be told as


```python
time_name = "Time"
```

Since the data are not in SI units, we need extra identifiers for the data after unit conversion. Let's give them endings in their variables as

```python
# string endings for readings after converting to SI values
k_str_end = "_K"
vol_str_end = "_m3s"
```

We also need to define the tolerance for steady state data. They are defined as the rate of change of temperature, power consumption and volumetric flow rates as

```python
t_toler = 0.05  # temperature change in K/min
rel_toler = 0.1/100.0  # relative change per minute
```

Don't forget the uncertainty from the manufacturers. They are listed here as

```python
temp_F_uncer = 0.05  # temperature measurement in Rankine
vdot_rel_uncer = 0.01  # relative volumetric flow rate
kW_rel_uncer = 0.015  # relative power consumption
```

Now set an empty list to store all DataFrames for multiple steady state data points.

```python
all_ss_df_options = []
```

And prepare the names of variables in SI units as

```python
temp_mea_si_names = [
    name+k_str_end for name in [
        tei_name, teo_name, tvdot_name
    ]+other_t_names
]
vdot_mea_si_names = [
    name+vol_str_end for name in [vdot_name]+other_vdot_names
]
```

# File Input and Data Verification

First, let's read the file.

```python
df_option = dread.read_csv_option(filepath)
```

Notice that the variable df_option is not a pandas DataFrame() but an OptionalVariable() defined in misc_func.py.

```python
type(df_option)
```

This is designed to help checking if the calculation or function processing the variable is correct. If it is correct, the pandas Dataframe() stored can be obained by

```python
df_option.get()
```

And if you try to check its error, it will return

```python
df_option.getError()
```

And the boolean output from it will be

```python
bool(df_option)
```

However, if the function process has failed, bool(df_option) will return False and df_option.getError() will give you the error encountered in the function.

Now let's set some filters to avoid errorenous data in the DataFrame() by getting data with above-freezing water temperature and positive flow rates only.

```python
df_option.set(df_option.get()[df_option.get()[tei_name] > 32.0])
df_option.set(df_option.get()[df_option.get()[teo_name] > 32.0])
df_option.set(df_option.get()[df_option.get()[tvdot_name] > 32.0])
df_option.set(df_option.get()[df_option.get()[vdot_name] > 0.0])
```

You can see that you can only assign the variables to misc.OptionalVariable() by using the function .set().

# Data Formatting

Now let's make the data easier to process by formatting it before getting the steady state data. First, let's convert the units of the temperature and flow rate readings to SI units.

```python
df_option = dread.unit_convert(df_option, [tei_name, teo_name, tvdot_name, tset_name]+other_t_names)
df_option = dread.unit_convert(df_option, [vdot_name]+other_vdot_names, 'GPM', 'm3s')
```

Now let's check the pandas DataFrame again.

```python
df_option.get()
```

You can see new columns with data in SI units now on the right.

Next, let's put the information about the name of the column indicating time and the fault information to the DataFrame(). This is done by appending a new attribute called 'details' to the pandas DataFrame() as a dictionary.

```python
df_option = dread.time_indication(df_option, time_name)
df_option = dread.fault_allocation(df_option, fault_type, fault_level)
```

Now let's check the 'details' attribute.

```python
df_option.get().details
```

You can see that the information is passed to the DataFrame() inside the variable df_option.

# Getting Steady State Data Points

To begin, let's calculate the average rate of change of readings of temperature in 15 minutes at each time instant.

```python
df_option = dread.cal_for_ss(df_option, 'abs_slope', col_names=temp_mea_si_names, ss_time=ss_time)
```

Now let's check the slope values.

```python
df_option.get().TEI_K_abs_slope[0].get()
```

```python
df_option.get().TEI_K_abs_slope[900].get()
```

You can see that the first time instant in the data does not have any slope calculated because the experiment had not started for 15 minutes. The slope values are only calculated at later time instants when enough data are obtained according to the value in the argument ss_time.

Let's calculate the average rates of change of flow rates and power consumption relative to their magnitude in 15 minutes as follows.

```python
df_option = dread.cal_for_ss(df_option, 'rel_slope', col_names=vdot_mea_si_names+[kW_name], ss_time=ss_time)
```

Now we have all the values to extract steady state data. In this case, a 15 minute operation is a steady state data point when it is 15 minutes away from other steady state data points and its change of temperature, power consumption and flow rates in 15 minutes are within the tolerances indicated above. The code to run it is

```python
ss_df_options = dread.ss_identifier(
    df_option, col_names=temp_mea_si_names+vdot_mea_si_names+[
        kW_name, tset_name+k_str_end
    ],
    abs_slope_cols=temp_mea_si_names,
    rel_slope_cols=vdot_mea_si_names+[kW_name],
    abs_slope_thres=[t_toler for x in range(len(temp_mea_si_names))],
    rel_slope_thres=[
        rel_toler for x in range(len(vdot_mea_si_names+[kW_name]))
    ], deltaT=ss_time, ss_time=ss_time
)
```

Now let's check how many steady state data points we have.

```python
len(ss_df_options)
```

You can see that only 8 15-minute operation period in the data file 'test_data.csv' satisfy our steady state requirement. Their DataFrames() can be obtained as

```python
ss_df_options[0].get()
```

You can see that we do not have all columns specified in the mother DataFrame() object df_option.get(). That's because we have specified which columns we want in the argument col_names in the function dread.ss_identifier().

# Data Analysis

Now let's continue our demonstration with one of the DataFrames(), ss_df_options[0].get().

Before our analysis, we want to append the uncertainty values to the DataFrame(). This is particularly important for sensors with uncertainty defined by a relative uncertainty.

```python
for col_name in temp_mea_si_names:
    ss_df_options[0] = ds.append_uncer_to_df(
        ss_df_options[0], col_name, misc.r2k(temp_F_uncer), 0.0
    )
for col_name in vdot_mea_si_names:
    ss_df_options[0] = ds.append_uncer_to_df(
        ss_df_options[0], col_name, 0.0, vdot_rel_uncer
    )
ss_df_options[0] = ds.append_uncer_to_df(
    ss_df_options[0], kW_name, 0.0, kW_rel_uncer
)
```

And now we have columns of uncertainties in our DataFrame() as OptionalVariables().

```python
ss_df_options[0].get()
```

Now let's get a summary of temperature, flow rate and power consumption readings by taking their means in the data with uncertainties. alpha is used here to indicate what confidence interval we should use to account for unaccounted effects such as environmental noise statistically.

```python
ss_df_options[0] = ds.data_mean_cal(ss_df_options[0], temp_mea_si_names+vdot_mea_si_names+[
    kW_name, tset_name+k_str_end
], alpha=alpha)
```

If we check the 'details' attribute again, we will have a list of mean values of readings and the rates of change in the steady state data points.

```python
ss_df_options[0].get().details
```

We can use the means to calculate a heat transfer rate by

```python
ss_df_options[0] = ds.cal_q_from_sample_result(
    ss_df_options[0], vdot_name+vol_str_end, tvdot_name+k_str_end,
    teo_name+k_str_end, tei_name+k_str_end, hx_name=hx_name
)
```

The function ds.cal_q_from_sample_result() only needs the column names to identify where the mean values are as their names are generated in fixed patterns. The resultant heat transfer rate in W is

```python
ss_df_options[0].get().details['q_evap_mean_obs']
```

Its negative sign indicates that the heat transfer rate points towards the system (heat exchanger). And its uncertainty is

```python
ss_df_options[0].get().details['q_uncer_evap_mean_obs']
```

We can also calculates the instantaneous heat transfer rate at each time instant and taking their means to get an estimate of instantaneous heat transfer rate without assuming rates of change of temperature and mass flow rates to be zero throughout the entire steady state period. This is achieved by

```python
ss_df_options[0] = ds.cal_mdotdeltah_per_time(
    ss_df_options[0], vdot_name+vol_str_end, tvdot_name+k_str_end,
    teo_name+k_str_end, tei_name+k_str_end, hx_name=hx_name
)
```

```python
ss_df_options[0] = ds.cal_q_from_ind_mea(ss_df_options[0], deltat=0, hx_name=hx_name, alpha=alpha)
```

```python
ss_df_options[0].get().details['q_evap_mean_inst']
```

```python
ss_df_options[0].get().details['q_uncer_evap_mean_inst']
```

If you want to get an estimate of mean heat transfer rate in every 30 seconds, the instantaneous heat transfer rates in the DataFrame() object can be integrated every 30s and their means will be able to give you this estimate. However, you reduce the number of samples of heat transfer rates per data point, and this may increase its noise estimated due to other factors.

```python
ss_df_options[0] = ds.cal_q_from_ind_mea(ss_df_options[0], deltat=30, hx_name=hx_name, alpha=alpha)
```

```python
ss_df_options[0].get().details['q_evap_mean_30s']
```

```python
ss_df_options[0].get().details['q_uncer_evap_mean_30s']
```

# File Output

After calculating the results for each steady state data points, you can append them to a list. As an example,

```python
all_ss_df_options.append(ss_df_options[0])
```

And you can pass the list to a ds.print_data() function to output it to a csv file as

```python
ds.print_data(
    all_ss_df_options, './test_data_summary.csv',
    detail_names=[
        'TEI_K_mean', 'TEO_K_mean',
        'TCI_K_mean', 'TCO_K_mean', 'TWE_set_K_mean',
        'FWE_m3s_mean', 'FWC_m3s_mean',
        'fault_type', 'fault_level',
        'kW_mean', 'q_evap_mean_inst',
        'TEI_K_mean', 'TEO_K_mean',
        'TCI_K_mean_uncer', 'TCO_K_mean_uncer',
        'FWE_m3s_mean_uncer', 'FWC_m3s_mean_uncer',
        'kW_mean_uncer', 'q_uncer_evap_mean_inst',
        'TEI_K_abs_slope', 'TEO_K_abs_slope',
        'TCI_K_abs_slope', 'TCO_K_abs_slope',
        'FWE_m3s_rel_slope', 'FWC_m3s_rel_slope',
        'kW_rel_slope'
    ]
)
```

The names in the argument details_names show what information in the 'details' attribute is passed to the output file. Now you can check the file for the summarized calculation!
